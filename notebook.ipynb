```python
# read data and import library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

data = pd.read_csv('magical_creatures_by_kingdom.csv')
```

**The Vibrant Realms of Numeria**

The magical kingdoms of Numeria each flourish with their unique blend of creatures. The Enchanted Forest, a central realm, remains the heart of fairy life with 40 individuals, surrounded by elves and goblins. Twilight Plains and Whispering Woods similarly showcase a rich variety of creatures, from unicorns to dragons. The Crystal Caves and Mystic Mountains maintain balanced mixes of trolls and goblins, while Shimmering Shores, despite its smaller size, retains a harmonious diversity with a spread of all creature types. Together, these regions paint a vivid picture of Numeriaâ€™s magical diversity and balance.

```python
# Create the Sunburst Chart
fig = px.sunburst(
    data,
    path=['Kingdom', 'CreatureType'],  # Define the hierarchical structure
    values='Number_of_Creatures',       # Define the values for each segment
    title='Distribution of Magical Creatures in Numeria',
    color='Number_of_Creatures',        # Use the number of creatures to define colors
    color_continuous_scale=px.colors.sequential.Plasma  # Set a beautiful color scale
)

# Set the size of the chart
fig.update_layout(
    width=800,  # Set the width of the chart
    height=600  # Set the height of the chart
)

# Show the figure
fig.show()
```

**The Mystical Balance of Numeria**

In the enchanted land of Numeria, each kingdom harbored its unique collection of magical creatures. The Enchanted Forest, with 20.5% of all creatures, was the heart of magic. Close behind, the Twilight Plains housed 19.6%, thriving with fairies and goblins under serene skies. Whispering Woods, at 18.8%, echoed with the secrets of elves and trolls, while the Crystal Caves and Mystic Mountains each sheltered 15.2% of the magical populace. The quietest of all, Shimmering Shores, held just 10.7%, hinting at unexplored mysteries awaiting discovery.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('magical_creatures_by_kingdom.csv')
pie_df = data.groupby("Kingdom").sum('Number_of_Creatures').reset_index()

# Creating the Pie Chart
plt.figure(figsize=(8, 6))
plt.pie(pie_df['Number_of_Creatures'],
        labels=pie_df['Kingdom'], 
        autopct='%1.1f%%', 
        startangle=140
)

# Adding title
plt.title('Distribution of Magical Creatures Across Kingdoms')

# Show the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.show()
```

**The Diverse Realms of Numeria**

The lands of Numeria teem with magical diversity. The Enchanted Forest emerges as the most populated realm, home to over a hundred creatures with a vibrant blend of fairies, goblins, and the largest population of elves. Twilight Plains and Whispering Woods closely follow, with balanced distributions of mystical beings, highlighting their harmonious environments.

The Mystic Mountains, with a high concentration of trolls and goblins, stand rugged yet lively, while the Crystal Caves maintain a moderate creature count, primarily elves and fairies. The quietest of all is Shimmering Shores, hosting a smaller number, yet rich in unicorns, hinting at its serene and unexplored nature.

```python
# Pivot the DataFrame to prepare for plotting
bar_df = data.pivot(index='Kingdom', columns='CreatureType', values='Number_of_Creatures')
bar_df = bar_df.fillna(0)

# Plotting
colors = plt.cm.plasma(np.linspace(0, 1, len(bar_df.columns)))
bar_df.plot(kind='bar', stacked=True, figsize=(10, 6), color=colors)

# Add title and labels
plt.title('Distribution of Magical Creatures Across Kingdoms', fontsize=16)
plt.xlabel('Kingdom', fontsize=14)
plt.ylabel('Number of Creatures', fontsize=14)
plt.xticks(rotation=45) 
plt.legend(title='Creature Type', bbox_to_anchor = (1.2, 0.6), loc='center right')

# Show the plot
plt.tight_layout()
plt.show()
```

**Numeriaâ€™s Diverse Creature Landscape**

The kingdoms of Numeria each nurture a unique mix of magical beings. The Enchanted Forest stands out with the highest concentration of fairies, numbering 40, making it a central haven for them. In the Twilight Plains, elves thrive with a strong presence of 30, while Whispering Woods is known for its dense population of both fairies and goblins. Mystic Mountains house a significant number of trolls, while Crystal Caves maintain a balanced presence of dragons and goblins. Shimmering Shores, though quieter, shines with its diversity in fairies and unicorns, showcasing Numeriaâ€™s rich and varied magical tapestry.

```python
# Pivot the DataFrame to prepare for plotting
heatmap_df = data.pivot(index='Kingdom', columns='CreatureType', values='Number_of_Creatures')
heatmap_df = heatmap_df.fillna(0)

# Create heatmap with Plasma colormap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_df, annot=True, cmap='coolwarm', fmt='d')

# Adding title and labels
plt.title('Distribution of Magical Creatures Across Kingdoms', fontsize=16)
plt.xlabel('Creature Type', fontsize=12)
plt.ylabel('Kingdom', fontsize=12)

# Show the plot
plt.show()
```
